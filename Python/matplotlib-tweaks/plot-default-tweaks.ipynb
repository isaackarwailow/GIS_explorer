```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rcParams, cycler
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/blue-yonder/pydse/master/pydse/data/international-airline-passengers.csv', sep=';')
```

```python
_ = plt.plot(df['Passengers'])
```

```python
# rcParams can be used to change the default plot size. This is so that tweaking needs not be chart related/ chart agnostic. That is, it can apply to any chart in any notebook
# restarting code execution for plots also don't change the size as default parameter is changed
rcParams['figure.figsize'] = 16, 8
rcParams['axes.spines.top'] = False
rcParams['axes.spines.right'] = False
rcParams['lines.linewidth'] = 2.5
rcParams['axes.prop_cycle'] = cycler(color=['#424242'])
rcParams['xtick.labelsize'] = 'xx-large'
rcParams['ytick.labelsize'] = 'xx-large'
```
